```python
# importation des fichiers csv
population = pd.read_csv('population.csv')
sous_nutrition = pd.read_csv('sous_nutrition.csv')
dispo_alimentaire = pd.read_csv('dispo_alimentaire.csv')
aide_alimentaire = pd.read_csv('aide_alimentaire.csv')
```

## Fichier  " Population"

```python
# v√©rification de l'importation
population.head()
```

```python
# renomme les colonnes selectionn√©es
population = population.rename(columns={'Zone': 'Pays', 'Valeur': 'Population'})
```

```python
# conversion de milliers d'habitant en haitant
population['Population'] *= 1000
population.head()
```

```python
# cr√©ation d'un dataFrame 
population_2017 = population[population['Ann√©e'] == 2017]
population_2017.head()
```

## Fichier "Sous_nutrition"

```python
sous_nutrition.head()
```

```python
sous_nutrition = sous_nutrition.rename(columns={'Zone': 'Pays', 'Valeur': 'Sous_nut'})
```

```python
# v√©rification et comptage des donn√©es manquantes ou nulles
sous_nutrition.isnull().sum()
```

```python
# change le type de donn√©es en type num√©rique 
sous_nutrition['Sous_nut']=pd.to_numeric(sous_nutrition['Sous_nut'],errors='coerce')
```

```python
# remplace les donn√©es nulle ou manquante par 0
sous_nutrition.fillna(0, inplace=True)
sous_nutrition
```

```python
# conversion de million d'habitant en habitant
sous_nutrition['Sous_nut'] *= 1000000
sous_nutrition.head()
```

```python
sous_nutrition_2017 = sous_nutrition[sous_nutrition['Ann√©e'] == '2016-2018'].copy()
sous_nutrition_2017.head()
```

## Fichier 'Dispo_alimentaire'

```python
dispo_alimentaire.head()
```

```python
dispo_alimentaire = dispo_alimentaire.rename(columns={'Zone': 'Pays'})
```

```python
dispo_alimentaire.isnull().sum()
```

```python
dispo_alimentaire.fillna(0, inplace=True)
dispo_alimentaire.head()
```

```python
# conversion avec une boucle for de milliers de tonnes en kg
col_liste = ['Aliments pour animaux', 'Disponibilit√© int√©rieure', 'Exportations - Quantit√©',
                        'Importations - Quantit√©', 'Nourriture', 'Pertes', 'Production',
                        'Semences', 'Traitement', 'Variation de stock', 'Autres Utilisations']

for elt in col_liste:
    dispo_alimentaire[elt] *= 1000000

dispo_alimentaire.head()
```

```python
# conversion de la tonne en kg
aide_alimentaire['Valeur'] *= 1000
aide_alimentaire.head()
```

## Demandes de Marc

### 1- La proportion de personnes en √©tat de sous_nutrition

```python
# cr√©ation d'un dataFrame pour l'ann√© 2017
prop_sous_nutrition = pd.merge(population_2017, sous_nutrition_2017,
                              on='Pays')

prop_sous_nutrition_2017 = prop_sous_nutrition[['Pays', 'Population', 'Sous_nut']]

prop_sous_nutrition_2017.head()
```

```python
prop_sous_nut_2017= round(prop_sous_nutrition_2017['Sous_nut'].sum()/(prop_sous_nutrition_2017['Population'].sum())*100,2)
prop_sous_nut_2017
print('La proportion de personnes en √©tat de sous nutrition en 2017 √©tait de', prop_sous_nut_2017, '%')
```

### *2- Nombre de personnes th√©orique pouvant √™tre nourries*

```python
nourriture_dispo_2017 = pd.merge(dispo_alimentaire, population_2017,
                               on='Pays')
# nourriture_dispo_2017 = nourriture_dispo[nourriture_dispo['Ann√©e'] == 2017]
nourriture_dispo_2017.head()
```

```python
nourriture_dispo_2017 = nourriture_dispo_2017[['Population', 'Disponibilit√© alimentaire (Kcal/personne/jour)', 'Origine', 'Pays']]
nourriture_dispo_2017.head()
```

```python
# cr√©ation d'un dataFrame et d'une colonne
nourriture_dispo_2017['Kcalj']= nourriture_dispo_2017['Disponibilit√© alimentaire (Kcal/personne/jour)']*nourriture_dispo_2017['Population']*365
dispo_2017 = round((nourriture_dispo_2017['Kcalj'].sum()/(2500*365)))

# somme de la totalit√©
pop_mondial_2017 = population_2017['Population'].sum()

# calcul de la nourriture disponible par rapport √† la population mondial
nb_theorique_2017 = round((dispo_2017/pop_mondial_2017)*100,2)


print('Le nombre th√©orique de personnes pouvant √™tre nourries en 2017 √©tait de', dispo_2017)
print('La proportion th√©orique de personnes pouvant √™tre nourries en 2017 √©tait de', nb_theorique_2017, '%')
```

```python
comptage_dispo_pop = nourriture_dispo_2017['Origine'].value_counts()
sns.barplot(x=comptage_dispo_pop.index, y=comptage_dispo_pop.values);
```

### *3- La disponibilit√© alimentaire des produits v√©g√©taux*

```python
# cr√©ation et selection d'une donn√©e pr√©cise dans un dataFrame
nourriture_dispo_2017 = nourriture_dispo_2017[nourriture_dispo_2017['Origine']=='vegetale']

# somme de la colonne selectionn√©
dispo_vegetaux = (nourriture_dispo_2017['Kcalj'].sum()/(2500*365))

# calcul de la disponibilit√© v√©g√©tale par rapport √† la population mondial
produits_vegetaux_dispo = round((dispo_vegetaux/pop_mondial_2017)*100,2)

print('La disponibilit√© alimentaire des produits v√©g√©taux en 2017 √©tait de', dispo_vegetaux)
print('La proportion alimentaire des produits v√©g√©taux en 2017 √©tait de', produits_vegetaux_dispo, '%')
```

### *4 "Disponibilit√© int√©rieure"*

```python
dispo_interieure = pd.merge(dispo_alimentaire, population, on='Pays')

# selection par position avec la m√©thode iloc 
disponibilite_alimentaire = dispo_alimentaire.iloc[:,3:].sum()

total_dispo_interieure = dispo_alimentaire['Disponibilit√© int√©rieure'].sum()
ratio_animal = round(dispo_alimentaire['Aliments pour animaux'].sum()/total_dispo_interieure*100,2)
ratio_nourriture = round(dispo_alimentaire['Nourriture'].sum()/total_dispo_interieure*100,2)
ratio_pertes = round(dispo_alimentaire['Pertes'].sum()/total_dispo_interieure*100,2)

print('Le ratio de nourriture animal est de', ratio_animal, '%')
print('Le ratio de nourriture est de', ratio_nourriture, '%')
print('Le ratio de pertes est de', ratio_pertes, '%')
```

## Demandes de M√©lanie

### *1- Pays pour lesquels la proportion de personnes sous-aliment√©es est la plus forte en 2017*

```python
prop_pays_sous_alimentes_2017 = pd.merge(sous_nutrition_2017, population_2017,
                               on='Pays')

prop_pays_sous_alimentes_2017.head(15)
```

```python
prop_pays_sous_alimentes_2017['Proportion'] = round ((prop_pays_sous_alimentes_2017['Sous_nut']/ prop_pays_sous_alimentes_2017['Population'])*100,2)

# dataFrame group√© par le total par pays
prop_pays_sous_alimentes_2017 = prop_pays_sous_alimentes_2017[['Pays', 'Proportion']].groupby('Pays').sum()

# dataFrame trier par ordre d√©croissant
prop_pays_sous_alimentes_2017.sort_values(['Proportion'], ascending = [False]).head(15)
```

### *Recherches de Julien*

```python
dispo_cereales= ['Bl√© et produits', 'riz', 'orge et produits', 'Ma√Øs et produits', 'Seigle et produits', 'Avoine', 'Millet et produits', 'Sorgho et produits', 'C√©r√©ales,Autres']

# selection de la donn√©e avec la fonction .loc et filtre les valeurs donn√©es pour la colonne sp√©cifi√©
cereales = dispo_alimentaire.loc[dispo_alimentaire['Produit'].isin(dispo_cereales),:]

prop_alim_animal = round(cereales['Aliments pour animaux'].sum()/cereales['Disponibilit√© int√©rieure'].sum()*100,2)
prop_nourriture = round(cereales['Nourriture'].sum()/cereales['Disponibilit√© int√©rieure'].sum()*100,2)
print("La proportion d'aliments pour animaux est de", prop_alim_animal, '%')
print("La proportion de nourriture humaine est de", prop_nourriture, '%')
```
